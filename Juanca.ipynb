# Matriz como lista

```python
import numpy as np 

def encontrar_manzana(M):

    filas = len(M) # Numero de filas
    columnas = len(M[0]) # Numero columnas
    posiciones_ij = []
    Vecinos = []

    for i in range(filas): 
        for j in range(columnas): 


            if(filas!=1 and columnas!=1):
                if((i > 0 and i < filas - 1) and (j > 0 and j < columnas - 1)): # no en los bordes
                    vecinos = M[i][j+1] + M[i][j-1] + M[i+1][j] + M[i-1][j] + M[i-1][j-1] + M[i-1][j+1] + M[i+1][j-1] + M[i+1][j+1]
                elif(j == 0 and i != 0 and i != filas - 1): # primera columna, no esquinas
                    vecinos = M[i][j+1] + M[i+1][j] + M[i-1][j] + M[i-1][j+1] + M[i+1][j+1]
                elif(j== 0 and i == 0): # primera columna, esquina superior izquierda
                    vecinos = M[i][j+1] + M[i+1][j] + M[i+1][j+1]
                elif(j == 0 and i == filas - 1): # primera columna, esquina inferior izquierda
                    vecinos = M[i][j+1] + M[i-1][j] + M[i-1][j+1] 
                elif(j == columnas - 1 and i != 0 and i != filas - 1): # ultima columna, no esquinas
                    vecinos = M[i][j-1] + M[i+1][j] + M[i-1][j] + M[i+1][j-1] + M[i-1][j-1]
                elif(j == columnas - 1 and i == 0): # ultima columna, esquina superior derecha
                    vecinos = M[i][j-1] + M[i+1][j] + M[i+1][j-1]
                elif(j == columnas - 1 and i == filas - 1): # ultima columna, esquina inferior derecha
                    vecinos = M[i][j-1] + M[i-1][j] + M[i-1][j-1]
                elif(i == 0 and j != 0 and j != columnas - 1): # primera fila, no esquinas
                    vecinos = M[i][j+1] + M[i][j-1] + M[i+1][j] + M[i+1][j+1] + M[i+1][j-1]
                elif(i == 0 and j != 0 and j != columnas - 1): # ultima fila, no esquinas
                    vecinos = M[i][j+1] + M[i][j-1] + M[i-1][j] + M[i-1][j+1] + M[i-1][j-1]
               
               
            elif(filas == 1 and columnas != 1): # si la matriz es solo una fila con varias columnas
                if(i == 0 and j != 0 and j != columnas - 1): # primera fila, no esquinas
                    vecinos = M[i][j+1] + M[i][j-1] 
                elif(i == 0 and j == 0): # Esquina izquierda de la fila
                    vecinos = M[i][j+1]
                elif(i == 0 and j == columnas-1): # Esquina derecha de la fila
                    vecinos = M[i][j-1]

            elif(filas != 1 and columnas == 1): # si la matriz es solo uan columna con muchas filas
                if(j == 0 and i != 0 and i != filas - 1): # primera columna, no esquinas
                    vecinos = M[i+1][j] + M[i-1][j] 
                elif(i == 0 and j == 0): # Esquina superior de la columna
                    vecinos = M[i+1][j]
                elif(i == filas -1 and j == 0): # Esquina inferior de la columna
                    vecinos = M[i-1][j]

            elif(filas == 1 and columnas == 1): # si la matriz es solo de una columna y una fila
                vecinos = 0 # no hay vecinos


           
            posiciones_ij.append([i,j]) # Guardo en la lista "posiciones_ij" todas las posiciones (i,j)
            Vecinos.append(vecinos) # Guardo en la lista "Vecinos" el numero de vecinos correspondiente a esa posicion (i,j)
            #print("Lista numero de vecinos menos o igual a 8:", Vecinos)
            #print("Posiciones:", posiciones_ij)


    maximo_vecinos = np.max(Vecinos) #Valor maximo de vecinos de la lista "Vecinos"
    posiciones_ij_maximo_vecinos = [] #Posicion (i,j) donde se tiene el maximo numero de vecinos
    #print("Maximo Vecinos: ",maximo_vecinos)

    for i in range(len(Vecinos)):
        if(Vecinos[i] == maximo_vecinos):
            posiciones_ij_maximo_vecinos.append(posiciones_ij[i]) # De todas las posiciones (i,j), guardo solo aquellas donde el numero de vecinos sea el maximo

    #print(posiciones_ij_maximo_vecinos)


    lista_distancias_al_origen = []

    if( len(posiciones_ij_maximo_vecinos) > 1 ): # Caso en el que "posiciones_ij_maximo_vecinos" tenga mas de una tupla
        
        for tupla in posiciones_ij_maximo_vecinos: # Recorremos los elementos de la lista directamente
            distancia_al_origen = np.sqrt( (tupla[1] - 0)**2 + (tupla[0] - 0)**2 ) # Distancia cartesiana desde el origen: d = raiz( (j-0)^2 + (i-0)^2 ). Nota: j=x, i=y 
            lista_distancias_al_origen.append(distancia_al_origen) #Guardamos en la lista "lista_distancias_al_origen" todas las distancias al origen de las tuplas de la lista "posiciones_ij_maximo_vecinos"


        minimo_distancias = np.min(lista_distancias_al_origen) #Valor minimo de distancia de la lista "lista_distancias_al_origen"
        posiciones_ij_minima_distancia = [] #Posicion (i,j) donde se tiene la minima distancia al origen

        for i in range(len(lista_distancias_al_origen)):
            if(lista_distancias_al_origen[i] == minimo_distancias):
                posiciones_ij_minima_distancia.append(posiciones_ij_maximo_vecinos[i]) # De todas las posiciones (i,j) donde se tiene el maximo numero de vecinos, guardo solo aquellas cuya distancia al origen es minima en una lista llamada "posiciones_ij_minima_distancia"


        lista_distancias_a_x_igual_0 = []

        if( len(posiciones_ij_minima_distancia) > 1): # Caso donde la lista "posiciones_ij_minima_distancia" tiene varios elementos

            for tupla in posiciones_ij_minima_distancia: # Recorremos los elementos de la lista directamente
                distancia_a_x_igual_0 = abs(tupla[1] - 0) # Distancia en el eje x entre la posicion en x de la tupla y x=0. Recordar que j=x.
                lista_distancias_a_x_igual_0.append(distancia_a_x_igual_0)

            minimo_distancias_a_x_igual_0 = np.min(lista_distancias_a_x_igual_0)
            posiciones_ij_minima_distancia_a_x_igual_0 = []

            for i in range(len(lista_distancias_a_x_igual_0)):
                if(lista_distancias_a_x_igual_0[i] == minimo_distancias_a_x_igual_0):
                    posiciones_ij_minima_distancia_a_x_igual_0.append(posiciones_ij_minima_distancia[i]) # De todas las posiciones (i,j) donde el numero de vecinos es maximo y la distancia es minima, guardo solo aquellas cuya distancia a x=0 es minima en una lista llamada "posiciones_ij_minima_distancia_a_x_igual_0"

            tupla = posiciones_ij_minima_distancia_a_x_igual_0[0]

        else: # Caso donde "posiciones_ij_minima_distancia" tiene solo una tupla

            tupla = posiciones_ij_minima_distancia[0]

    else:
        tupla = posiciones_ij_maximo_vecinos[0] # En este caso "posiciones_ij_maximo_vecinos" solo tiene una tupla, que es el elemento [0] de "posiciones_ij_maximo vecinos"

    return tupla 


#Matriz = [[4,0,1,2],[6,5,4,4],[4,0,4,1],[7,7,4,5]]
Matriz = [[2]]
print(Matriz) #Se imprime la matriz
print("\n")
print(encontrar_manzana(Matriz)) #Se ejecuta la funcion para la matriz
```
